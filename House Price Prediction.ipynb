```python
import pandas as pd
data = pd.read_csv("C:\\kc_house_data.csv")
data.sample(5)
```

```python
data.drop(["date", "id"], axis=1, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
corr = data.corr()
sns.heatmap(corr, annot=True, cmap="Blues", fmt=".1f", linewidth=1)
plt.title("Correlation", fontsize=17)
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X = data.drop(["price", "zipcode"], axis=1)
y = data["price"]
```

```python
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
lrs = model.score(X_test, y_test)
```

```python
from sklearn.linear_model import Ridge
scores = []
for i in range(1,101):
    r = Ridge(alpha=i)
    r.fit(X_train, y_train)
    score = r.score(X_test, y_test)
    scores.append(score)
```

```python
plt.plot(range(1,101), scores)
```

```python
rs = max(scores)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
rfrs = rfr.score(X_test, y_test)
```

```python
index = ["linear", "ridge", "random forest"]
values = [lrs, rs, rfrs]
sns.barplot(x=index, y=values, alpha=0.7)
plt.title("Accuracy Scores of Models", fontsize=15)
plt.ylabel("Accuracy Score", fontsize=15)
plt.xlabel("Models", fontsize=17)
lrs, rs, rfrs
```
